```python
import pandas as pd
import numpy as np
import os
pd.set_option("precision", 4)
pd.options.display.float_format = '${:.4f}'.format
```

```python
csv = [os.path.join("./results/", f) for f in os.listdir("./results/") if f[-4:] == ".csv"]
csv
```

```python
name_map = {
    "arnoldi": "Arnoldi",
    "glv": "GloVe",
    "w2v": "word2vec"
}
df = pd.DataFrame()
for f in sorted(csv):
    params = f.split(sep="_")[1:]
    params[-1] = params[-1][:-4]
    params[0] = name_map[params[0]]
#     if params[1] == "500":
#         continue
    row = pd.read_csv(f)
    row["method"] = params[0]
    row["dim"] = params[1]
    row["win_size"] = params[2]
    df = df.append(row)

df = df[df.columns[1:]]
df.set_index(["method", "dim", "win_size"], inplace=True)


def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

df.style.apply(highlight_max)
```
